```python
import pandas as pd

# Open data.csv in a dataframe
df = pd.read_csv('hacksc/data/data.csv')

# Print the head of the dataframe
print(df.head())
```

```python
# Rename first unnamed column to id
df.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
```

```python
# Print the columns of the dataframe
for column in df.columns:
    print(column, df[column].dtype)
```

```python
for index, row in df.iterrows():
    if index >= 5000:
        break
    if row['cc_armswelling'] != 0.0:
        print(row['cc_armswelling'])
```

```python
for index, row in df.iterrows():
    if index >= 5000:
        break
    cc_columns = [column for column in df.columns if column.startswith('cc_')]
    cc_values = [row[column] for column in cc_columns if row[column] == 1.0]
    if len(cc_values) > 1:
        print(cc_values)
```

```python
# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Print the first few rows of the dataframe
print(df.head())
```

```python
# Reset the option to its default value
pd.reset_option('display.max_columns')
```

```python
# Take 10 rows out of the df and make a new df with them
mockup_df = df.head(20)
mockup_df.drop(columns=['id'], inplace=True)
```

```python
# All chief complaint columns
cc_columns = [col for col in mockup_df.columns if col.startswith('cc_')]

# Column called chief_complaint with an array of the chief complaints for each patient row
def get_chief_complaints(row):
    return [col.replace('cc_', '') for col in cc_columns if row[col] == 1.0]

mockup_df['chief_complaint'] = mockup_df.apply(get_chief_complaints, axis=1)

unique_complaints = set([complaint for complaints in mockup_df['chief_complaint'] for complaint in complaints])

# Constructing an enum for all possible chief complaints
from enum import Enum

ChiefComplaint = Enum('ChiefComplaint', {complaint.upper().replace(' ', '_'): complaint for complaint in unique_complaints})

for complaint in ChiefComplaint:
    print(complaint)
```

```python
# All meds columns
meds_columns = [col for col in mockup_df.columns if col.startswith('meds_')]

# Column called meds with an array of the meds for each patient row
def get_meds(row):
    return [col.replace('meds_', '') for col in meds_columns if row[col] == 1.0]

mockup_df['meds'] = mockup_df.apply(get_meds, axis=1)

unique_meds = set([med for meds in mockup_df['meds'] for med in meds])

# Constructing an enum for all possible chief complaints
from enum import Enum

Meds = Enum('Meds', {med.upper().replace(' ', '_'): med for med in unique_meds})

for med in Meds:
    print(med)
```

```python
import pandas as pd
import numpy as np

condition_columns = mockup_df.loc[:, '2ndarymalig':'whtblooddx']

def get_conditions(row):
    return row.index[row == 1.0].tolist()

mockup_df['conditions'] = condition_columns.apply(get_conditions, axis=1)
```

```python
from enum import Enum

# Get all unique conditions
unique_conditions = np.unique(mockup_df['conditions'].explode().values)

# Create an Enum with these conditions
class Condition(Enum):
    pass

# Dynamically add the conditions to the Enum
for condition in unique_conditions:
    setattr(Condition, condition, condition)
```

```python
mockup_df.drop(df.loc[:, '2ndarymalig':'whtblooddx'].columns, axis=1, inplace=True)
mockup_df = mockup_df.loc[:,~mockup_df.columns.str.startswith('meds_')]
mockup_df = mockup_df.loc[:,~mockup_df.columns.str.startswith('cc_')]
```
